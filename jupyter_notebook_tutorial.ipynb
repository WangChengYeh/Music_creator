### ğŸ“‹ å„²å­˜æ ¼æ“ä½œæŒ‡å—ï¼š

**åœ¨å‘½ä»¤æ¨¡å¼ä¸‹ï¼ˆæŒ‰ Esc é€²å…¥ï¼‰ï¼š**
- `A`ï¼šåœ¨ä¸Šæ–¹æ’å…¥æ–°å„²å­˜æ ¼

### ğŸ’¡ å¯¦ç”¨æŠ€å·§ï¼š
- ä½¿ç”¨ `Shift + é»æ“Š` é¸æ“‡å¤šå€‹å„²å­˜æ ¼
- æ‹–æ‹½å„²å­˜æ ¼å·¦å´çš„æ¢ä¾†ç§»å‹•ä½ç½®
- ä½¿ç”¨ `Ctrl + /` ä¾†è¨»è§£/å–æ¶ˆè¨»è§£ç¨‹å¼ç¢¼

````python
# äº’å‹•å¼ç¯„ä¾‹ï¼šå˜—è©¦ä¿®æ”¹é€™å€‹å„²å­˜æ ¼
# æ‚¨å¯ä»¥ä¿®æ”¹ä¸‹é¢çš„åƒæ•¸ï¼Œç„¶å¾ŒæŒ‰ Shift+Enter é‡æ–°åŸ·è¡Œ

# å¯ä¿®æ”¹çš„åƒæ•¸
amplitude = 2.0  # æŒ¯å¹…
frequency = 1.0  # é »ç‡
phase = 0.0      # ç›¸ä½
points = 1000    # é»æ•¸

# ç”Ÿæˆæ­£å¼¦æ³¢æ•¸æ“š
x = np.linspace(0, 4*np.pi, points)
y = amplitude * np.sin(frequency * x + phase)

# ç¹ªè£½åœ–è¡¨
plt.figure(figsize=(12, 6))
plt.plot(x, y, 'b-', linewidth=2, label=f'y = {amplitude} Ã— sin({frequency}x + {phase})')
plt.title('å¯èª¿æ•´çš„æ­£å¼¦æ³¢', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.show()

print(f"ğŸ“ˆ ç•¶å‰åƒæ•¸ï¼šæŒ¯å¹…={amplitude}, é »ç‡={frequency}, ç›¸ä½={phase}")
print("ğŸ’¡ æç¤ºï¼šå˜—è©¦ä¿®æ”¹ä¸Šé¢çš„åƒæ•¸å€¼ï¼Œç„¶å¾Œé‡æ–°åŸ·è¡Œé€™å€‹å„²å­˜æ ¼ï¼")
````
